```python
import hail as hl
hl.init(default_reference='GRCh38')
```

```python
from bokeh.io import show, output_notebook
from bokeh.layouts import gridplot
output_notebook()
```

```python
mt_path = "gs://activestorage_amppd_mgrb/v1-2-amp-pd-mgrb.chr22.filtered.dbSNFP.dense.mt/"
```

```python
mt = hl.read_matrix_table(mt_path)
```

```python
mt.describe()
```

```python
mt.info.AS_QUALapprox.show()
```

```python
# function to plot histograms 
def plot_histogram(data, legend):
    p = hl.plot.histogram(data, legend=legend, title=f"{legend} Histogram")
    show(p)
```

```python
AS_QUALapprox = mt.aggregate_entries(hl.expr.aggregators.hist(mt.info.AS_QUALapprox, 0, 1000, 90))
plot_histogram(AS_QUALapprox, "AS_QUALapprox_hist")
```

```python
AS_QD = mt.aggregate_entries(hl.expr.aggregators.hist(mt.info.AS_QD, 0, 1000, 30))
plot_histogram(AS_QD, "AS_QD")
```

```python
QD = mt.aggregate_entries(hl.expr.aggregators.hist(mt.info.QD, 0, 1000, 30))
plot_histogram(QD, "QD")
```

```python
QUALapprox = mt.aggregate_entries(hl.expr.aggregators.hist(mt.info.QUALapprox, 0, 1000, 30))
plot_histogram(QUALapprox, "QUALapprox")
```

```python
variant_qc_gq_stats_mean = mt.aggregate_entries(hl.expr.aggregators.hist(mt.variant_qc.gq_stats.mean, 0, 160, 30))
plot_histogram(variant_qc_gq_stats_mean, "variant_qc_gq_stats_mean")
```

```python
variant_qc_gq_stats_min = mt.aggregate_entries(hl.expr.aggregators.hist(mt.variant_qc.gq_stats.min, 0, 1000, 30))
plot_histogram(variant_qc_gq_stats_min, "variant_qc_gq_stats_min")
```

```python
variant_qc_gq_stats_max = mt.aggregate_entries(hl.expr.aggregators.hist(mt.variant_qc.gq_stats.max, 0, 120, 30))
plot_histogram(variant_qc_gq_stats_max, "variant_qc_gq_stats_max")
```

```python
RGQ = mt.aggregate_entries(hl.expr.aggregators.hist(mt.RGQ, 0, 1000, 30))
plot_histogram(RGQ, "RGQ")
```

```python
GQ = mt.aggregate_entries(hl.expr.aggregators.hist(mt.GQ, 0, 1000, 30))
plot_histogram(GQ, "GQ")
```

```python
gvcf_info_QD = mt.aggregate_entries(hl.expr.aggregators.hist(mt.gvcf_info.QD, 0, 1000, 30))
plot_histogram(gvcf_info_QD, "gvcf_info_QD")
```
